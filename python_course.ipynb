```python
def fib(n):
     a, b = 0, 1
     while a<n:
         print(a, end=' ')
         a, b = b, a+b
     print()

def fib2(n):
     result = []
     a, b = 0, 1
     while a<n:
         result.append(a)
         a, b = b, a+b
     return result

fib(100)
fib2(1000)
```

```python
from math import factorial, exp

def e_x(x, terms):
    sum = 0
    for i in range(terms):
        sum += (x ** i) / factorial(i)
    return sum

# Reading x and terms from file
f = open("input.txt", "r")
x = float(f.readline().split('=')[1])
terms = int(f.readline().split('=')[1])
f.close()

# Calculating
approx = e_x(x, terms)
actual = exp(x)
error = actual - approx

# Writing results to file
f = open("output.txt", "w")
f.write("e^" + str(x) + " (for " + str(terms) + " terms) = " + str(approx) + "\n")
f.write("Actual value = " + str(actual) + "\n")
f.write("Truncation Error = " + str(error) + "\n")
f.close()
```
